```python
from pyspark.sql import SparkSession, DataFrame
from pyspark.sql.types import StringType, StructType, IntegerType, DateType, StructField
import pyspark.sql.functions as F
import pandas as pd
```

```python
spark = SparkSession.builder \
    .appName("dataStorageSystems") \
    .master("local") \
    .config("spark.submit.deployMode", "client") \
    .config("spark.executor.instances", "2") \
    .config("spark.executor.cores", "2") \
    .config("spark.executor.memory", "2g") \
    .config("spark.driver.memory", "2g") \
    .config("spark.driver.cores", "2") \
    .getOrCreate()
```

```python
spark
```

```python
df = spark.read.parquet('./data/')
```

```python
df.columns
```

<h1>3</h1>

```python
clean_df = df.filter(F.date_format(F.col('tpep_pickup_datetime'), 'yyyy-MM-dd') > '2025-01-01') \
    .filter(F.date_format(F.col('tpep_pickup_datetime'), 'yyyy-MM-dd') < '2025-06-30') \
    .filter(F.date_format(F.col('tpep_dropoff_datetime'), 'yyyy-MM-dd') > '2025-01-01') \
    .filter(F.date_format(F.col('tpep_dropoff_datetime'), 'yyyy-MM-dd') < '2025-06-30') \
    .filter(F.col('trip_distance') > 0) \
    .filter(F.col('passenger_count') > 0)
```

<h1>4</h1>

```python
hours = clean_df \
    .withColumn('tpep_pickup_hour', F.hour('tpep_pickup_datetime')) \
    .withColumn('tpep_dropoff_hour', F.hour('tpep_dropoff_datetime'))
```

<h1>5</h1>

```python
res = hours.select(
    [
        'tpep_pickup_datetime',
        'tpep_pickup_hour',
        'tpep_dropoff_datetime',
        'tpep_dropoff_hour',
        'passenger_count',
        'trip_distance',
        'PULocationID',
        'DOLocationID',
    ]
)
```
